```python
from os import listdir
from os.path import isfile, join
import csv
import numpy as np
```

```python
dataset_dict = dict()
dataset_dict[2] = "MNIST"
dataset_dict[20] = "SVHN"
dataset_dict[38] = "Fashion-MNIST"
dataset_dict[56] = "CIFAR-10"
```

```python
# Picking the appropriate columns to analyse

pick = []
for i in range(36):
    pick.append(i*2 + 2)

pick_updated = []
for i in pick:
    if (i % 18 < 7 and i % 18 > 0):
        pick_updated.append(i)
print (pick_updated)
```

```python
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'

```

```python
def print_data(rows, pick):
    key = 0
    for k, i in enumerate(pick):
        confs_dict = {}
        options = set()
        all_conf = []

        for row in rows:
            options.add(row[i-1])

        for option in list(options):
            confs_dict[option] = []

        for row in rows:
            confs_dict[row[i-1]].append(int(row[i]))
            all_conf.append(int(row[i]))
            

        if (i in dataset_dict.keys()):
            print ("Dataset under test: ", color.BOLD + dataset_dict[i] + color.END)
            print()
        
        print("Results Image #", int(key % 3 + 1))
        for option in list(options):
            
            print("Option: ", color.BLUE + option + color.END)
            print("%users choosing option:", color.BLUE + str(round(100*len(confs_dict[option])/len(rows), 2))+ "%" + color.END) 
            print("Average Confidence: ", color.BLUE + str(round(np.mean(confs_dict[option]), 2)) + color.END)
            print()
#             print(option, np.mean(confs_dict[option]), str(100*len(confs_dict[option])/len(rows)) + "%")
            
        print("Average Confidence for all inputs:", color.UNDERLINE + str(round(np.mean(all_conf), 2)) + color.END)
        
        key += 1
        
        if (i % 3 == 0): 
            print ("============================")
        else:
            print ("---------")
        print()
```

```python
# Extracting Data from the CSV file

rows = []
with open('User-Study.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        rows.append(row)
rows = rows[1:]
print("Number of respondents:", len(rows))
print("Number of columns:", len(rows[0]))
```

```python
print_data(rows, pick_updated)
```
